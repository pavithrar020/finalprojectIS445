```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
barbie = pd.read_csv(r'desktop/barbieverse.csv')

# Group by Character Name and count movies
character_movie_counts = barbie.groupby('Character Name')['No.  Movie'].nunique()

# Plotting
plt.figure(figsize=(8, 5))
character_movie_counts.plot(kind='hist', bins=range(1, character_movie_counts.max()+2), rwidth=0.8, color='magenta')
plt.title('Movies per Character in Barbie Movies')
plt.xlabel('Number of Movies')
plt.ylabel('Number of Characters')
plt.xticks(range(1, character_movie_counts.max()+1))
plt.show()
```

```python
# Count the number of speaking and non-speaking characters
speaking_counts = barbie['Speaking Voice'].notnull().sum()
non_speaking_counts = len(barbie) - speaking_counts

# Plotting
plt.figure(figsize=(6, 6))
plt.bar(['Speaking Characters', 'Non-Speaking Characters'], [speaking_counts, non_speaking_counts], color=['#FF69B4', '#87CEFA'])
plt.title('Number of Speaking vs. Non-Speaking Characters in Barbie Movies')
plt.ylabel('Number of Characters')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
barbie = pd.read_csv('desktop/barbieverse.csv')

# Filter out invalid gender data
valid_genders = barbie['Gender'].isin(['Female', 'Male'])
barbie_valid_gender = barbie[valid_genders]

# Count the number of characters by gender
gender_counts = barbie_valid_gender['Gender'].value_counts()

# Define custom Barbie colors
barbie_colors = ['#FF69B4', '#4682B4']  # Pink and blue

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, colors=barbie_colors, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Barbie Characters by Gender')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
