```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
barbie = pd.read_csv(r'desktop/barbieverse.csv')

# Line plot - number of characters in each movie
plt.figure(figsize=(8, 6))
movie_character_counts = barbie.groupby('No.  Movie')['No. Character'].count()
plt.plot(movie_character_counts.index, movie_character_counts.values, marker='o')
plt.title('Number of Characters in Each Movie')
plt.xlabel('Movie Number')
plt.ylabel('Number of Characters')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
barbie = pd.read_csv(r'desktop/barbieverse.csv')

# Pie chart - distribution of character races
race_counts = barbie['Race'].value_counts()

# Increase figure size
plt.figure(figsize=(8, 8))

# Plot pie chart with labels and percentages removed
patches, _ = plt.pie(race_counts, labels=None)

# Calculate percentages
percentages = [f'{race_counts[i] / race_counts.sum() * 100:.1f}%' for i in range(len(race_counts))]

# Add legend with labels and percentages
plt.legend(patches, labels=[f'{race} ({percentage})' for race, percentage in zip(race_counts.index, percentages)], loc="center left", bbox_to_anchor=(1, 0.5))

plt.title('Distribution of Character Races')
plt.show()
```

```python
# Create a new column combining eye and hair color
barbie['Eye_Hair_Color'] = barbie['Eyes'] + ' & ' + barbie['Hair']

# Count combinations
eye_hair_combinations = barbie['Eye_Hair_Color'].value_counts().head(10)  # top 10 most frequent combinations

# Plot
plt.figure(figsize=(12, 6))
eye_hair_combinations.plot(kind='bar', color='teal')
plt.title('Top 10 Most Frequent Eye and Hair Color Combinations')
plt.xlabel('Eye & Hair Color')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
